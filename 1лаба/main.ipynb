```python
import pandas as pd
import numpy as np
import scipy.stats as stats


hee_quan_df = pd.DataFrame([[None] * len(columns)] * len(columns),
                      index=columns,
                      columns=columns)

for index_x in range(len(columns)):
    col_x = columns[index_x]
    for index_y in range(len(columns)):
        col_y = columns[index_y]

        # Создаем таблицу сопряженности
        contingency_table = pd.crosstab(quan_data[col_x], quan_data[col_y])

        # Вычисляем статистику хи-квадрат
        chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

        n = np.sum(contingency_table)
        phi = np.sqrt(chi2 / n)
        
        # коэффициент Крамера
        cramer_v = np.sqrt(chi2 / (n * min(contingency_table.shape[0]-1, contingency_table.shape[1]-1)))

        # Записываем значения в DataFrame
        # hee_df._set_value(col_x, col_y, chi2)
        # hee_df._set_value(col_x, col_y, phi)
        hee_df._set_value(col_x, col_y, cramer_v)

hee_df.to_excel("hee_quan.xlsx",sheet_name='Крамер') 
```
