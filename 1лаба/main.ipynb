```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif

df1 = pd.read_excel('quantitative_data.xlsx')
X1 = df1["A4"]
X2 = df1["A5"]
Y = df1["A16"]
# Рассчитываем взаимную информацию между X1 и Y
mi_X1_Y = mutual_info_classif(X1.values.reshape(-1, 1), Y, discrete_features=True)
print(f"Взаимная информация между X1 и Y: {mi_X1_Y[0]}")

# Рассчитываем взаимную информацию между X2 и Y
mi_X2_Y = mutual_info_classif(X2.values.reshape(-1, 1), Y, discrete_features=True)
print(f"Взаимная информация между X2 и Y: {mi_X2_Y[0]}")
```

```python
def conditional_mutual_information(data, X:set, Y:set, Z:set, delta = 1):
        X = list(X); Y = list(Y); Z = list(Z)
        cmi = 0

        P_Z = data.groupby(Z).size()
        P_Z = P_Z/P_Z.sum()

        P_XZ = data.groupby(X + Z).size()
        P_XZ = P_XZ/P_XZ.sum()

        P_YZ = data.groupby(Y + Z).size()
        P_YZ = P_YZ/P_YZ.sum()

        P_XYZ = data.groupby(X + Y + Z).size()
        P_XYZ = P_XYZ/P_XYZ.sum()

        for ind in P_XYZ.index:
            x_ind = ind[:len(X)]
            y_ind = ind[len(X):len(X + Y)]
            z_ind = ind[len(X + Y):]

            xz_ind = x_ind + z_ind
            yz_ind = y_ind + z_ind
            xyz_ind = ind

            z_ind =  pd.MultiIndex.from_tuples([z_ind], names = Z) if len(Z) != 1 else pd.Index(z_ind, name = Z[0])
            xz_ind = pd.MultiIndex.from_tuples([xz_ind], names = X + Z)
            yz_ind = pd.MultiIndex.from_tuples([yz_ind], names = Y + Z)
            xyz_ind = pd.MultiIndex.from_tuples([xyz_ind], names = X + Y + Z)

            cmi += delta * P_XYZ[xyz_ind].item() * np.log2(P_Z[z_ind].item() * P_XYZ[xyz_ind].item() / (P_XZ[xz_ind].item() * P_YZ[yz_ind].item()))

        return cmi

data = df1
data['X'] = X1
data['Y'] = Y
data['Z'] = X2
pmi1 = conditional_mutual_information(data, {'X'}, {'Y'}, {'Z'})

data['X'] = X2
data['Y'] = Y
data['Z'] = X1
pmi2 = conditional_mutual_information(data, {'X'}, {'Y'}, {'Z'})
print(pmi1, pmi2)
```
